```python
# Download the helper library from https://www.twilio.com/docs/python/install
import os
from twilio.rest import Client
import urllib.request
import time
import webbot
from datetime import datetime


# Your Account Sid and Auth Token from twilio.com/console
# and set the environment variables. See http://twil.io/secure
account_sid = os.environ['TWILIO_ACCOUNT_SID']
auth_token = os.environ['TWILIO_AUTH_TOKEN']
client = Client(account_sid, auth_token)

def call_me():
    call = client.calls.create(
                            url='http://demo.twilio.com/docs/voice.xml',
                            to='+13524082858',
                            from_='+13237468454'
                        )

    print(call.sid)
    


b = webbot.Browser()
b.go_to("https://scheduling.lakecountyfl.gov/")
tries = 1
time.sleep(1)
original_source = b.get_page_source()
#write out the current source
text_file = open("First.html", "w")
text_file.write(original_source)
text_file.close()

while True:
    
    current_source = b.get_page_source()
    tries = tries + 1
    if current_source != original_source:        
        print ("Calling  you master after " + str(tries) + " tries!")
        call_me()
        now = datetime.now()
        text_file = open( now.strftime("%d/%m/%Y %H_%M_%S") + ".html", "w")
        text_file.write(current_source)
        text_file.close()
        original_source = current_source
        time.sleep(30)
    else:
        now = datetime.now()
        print ("Loaded page " + str(tries) + " times on " + now.strftime("%d/%m/%Y %H_%M_%S"))
        time.sleep(30)
```
